```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# Load the data from CSV file
file_path = "C:/Users/61433/Desktop/Project-1---Group-3/Dataset/cleaned_data.csv"
data_df = pd.read_csv(file_path)
data_df.head()
```

```python
# Convert 'ActivityDate' to datetime
data_df['ActivityDate'] = pd.to_datetime(data_df['ActivityDate'])
```

```python
# Scatter plot Total Steps vs Calories with regression line (Q1)
plt.figure(figsize=(10, 6))
plt.scatter(data_df['TotalSteps'], data_df['Calories'],marker="o", color="green",edgecolor="black",alpha=0.5, label='Data Points')
plt.title('Total Steps vs Calories Burned')
plt.xlabel('Total Steps')
plt.ylabel('Calories Burned')

# Perform linear regression
slope, intercept, r_value, p_value, std_err = stats.linregress(data_df['TotalSteps'], data_df['Calories'])
plt.plot(data_df['TotalSteps'], slope * data_df['TotalSteps'] + intercept, color='red', label='Regression Line')

plt.legend()
plt.grid(True)
plt.show()
```

scatter plot: This scatter plot illustrates that as the number of steps taken increases, so does the number of calories burned.

Relationship between TotalSteps and Calories burned:

Null Hypothesis (H): There is no significant relationship between TotalSteps and Calories burned.

Alternative Hypothesis (H): There is a significant positive relationship between TotalSteps and Calories burned, indicating that as TotalSteps increase, Calories burned also increase.



```python
# Scatter plots for each activity vs Calories (Q2)
activities = ['VeryActiveMinutes', 'FairlyActiveMinutes', 'LightlyActiveMinutes', 'SedentaryMinutes']

for activity in activities:
    plt.figure(figsize=(10, 6))
    plt.scatter(data_df[activity], data_df['Calories'],marker="s", color="blue",edgecolor="black", alpha=0.5)
    plt.title(f'{activity} vs Calories Burned')
    plt.xlabel(activity)
    plt.ylabel('Calories Burned')
    plt.grid(True)
    plt.show()
```

scatter plots: These scatter plots demonstrate that higher levels of activity lead to burning more calories in fewer minutes.

Contribution of each activity to calorie burn:

Null Hypothesis (H): All activities (Very Active, Fairly Active, Lightly Active, Sedentary) contribute equally to calorie burn.

Alternative Hypothesis (H): One or more specific activities (e.g., Very Active) contribute significantly more to calorie burn compared to others.


```python
# 3. Average number of calories burned per day (q3)
# Calculate total minutes for each activity type
total_very_active = data_df['VeryActiveMinutes'].sum()
total_fairly_active = data_df['FairlyActiveMinutes'].sum()
total_lightly_active = data_df['LightlyActiveMinutes'].sum()
total_sedentary = data_df['SedentaryMinutes'].sum()

# Pie chart
labels = ['Very Active', 'Fairly Active', 'Lightly Active', 'Sedentary']
sizes = [total_very_active, total_fairly_active, total_lightly_active, total_sedentary]
colors = ['#4CB140','#5752D1','#F4C145','#EC7A08']
explode = (0.1, 0, 0, 0)  # explode the 1st slice

plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, shadow=True, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Activity Minutes')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

pie chart: This pie chart illustrates the distribution of activity levels based on minutes: Sedentary activities account for 82%, Lightly Active for 15.3%, Fairly Active for 1.1%, and Very Active for 1.6%.

Percentage of activity levels based on activity minutes:

Null Hypothesis (H): Activity levels (Very Active, Fairly Active, Lightly Active, Sedentary) are evenly distributed based on activity minutes.

Alternative Hypothesis (H): There is a skewed distribution where one or more activity levels account for a significantly higher percentage of total activity minutes compared to others.
