```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
from scipy.stats import linregress
```

```python
# uncleaned data file path
dataset_path = "uncleaned merged data.csv"

# read data
uncleaned_data = pd.read_csv(dataset_path)

# convert the uncleaned data into a Pandas DataFrame
uncleaned_df = pd.DataFrame(uncleaned_data)

uncleaned_df.head()
```

```python
# checking the number of unique ID's
id_count = uncleaned_df["Id"].nunique()
id_count
```

```python
# find the duplicate rows by  ID, activity date and total distance
duplicate_df = uncleaned_df[uncleaned_df.duplicated(subset=["Id", "ActivityDate", "TotalSteps"])]

# Remove duplicates from the original data based on specific columns
cleaned_data = uncleaned_df.drop_duplicates(subset=['Id', 'ActivityDate', 'TotalSteps'])

# Save the duplicates and cleaned data to separate files
#duplicates_file_path =   # Update this path to your desired location
#cleaned_file_path = 'path/to/your/cleaned_data.csv'  # Update this path to your desired location

duplicate_df.to_csv("duplicates.csv", index=False)
cleaned_data.to_csv("cleaned_data.csv", index=False)
```
